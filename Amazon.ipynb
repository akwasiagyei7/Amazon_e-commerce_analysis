### Impoting the data 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('Amazon Sales.csv')
```

## Data Processing

```python
df.head()
```

```python
df.tail()
```

#### Checking missing values

```python
df.isnull().sum()
#df.dropna(subset=['rating', 'review_title', 'rating_count', 'review_content'], inplace=True)
```

rating count has two rows that contains NA so we will remove them

```python
# Droping N/As and duplicates
df.dropna(subset=['rating', 'rating_count'], inplace=True)
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
# Checking data types
df.dtypes
```

#### Data type conversion to numeric

```python
df['discounted_price'] = df['discounted_price'].replace('[^0-9.]', '', regex=True).astype(float)
df['actual_price'] = df['actual_price'].replace('[^0-9.]', '', regex=True).astype(float)
df['discount_percentage'] = df['discount_percentage'].replace('%', '', regex=True).astype(float)
df['rating'] = pd.to_numeric(df['rating'].astype(str).str.replace(',', ''), errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'].astype(str).str.replace(',', ''), errors='coerce')
```

### Exploratory Data Analysis
#### Summary Statistics

```python
# Summary statistics
df.describe()
```

#### Distribution of Ratings

```python
sns.histplot(df['rating'], bins=10, kde=True)
plt.title('Distribtuion of Product Ratings')
plt.show()
```

#### Top Performing Products by Rating

```python
top_products = df.groupby('product_name')['rating'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(8, 5))
sns.barplot(y=top_products.index, x=top_products.values, hue=top_products.index, palette='viridis', legend=False)
plt.xlabel("Average Rating")
plt.ylabel("Product Name")
plt.title("Top 10 Products by Rating")
plt.show
```

#### Price vs. Rating Analysis
The scatterplot shows the relationship between discounted price and rating

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='discounted_price', y='rating', data=df, alpha=0.6)
plt.xscale("log")
plt.xlabel("Actual Price (log scale)")
plt.ylabel("Rating")
plt.title('Price vs. Rating')
plt.show
```

Boxplot to compare rating across different price range

```python
df['price_range'] = pd.cut(df['discounted_price'], bins=[0, 1000, 5000, 10000, 20000, 50000], 
                           labels=['0-1K', '1K-5K', '5K-10K', '10K-20K', '20K-50K'])
sns.boxplot(x='price_range', y='rating', data=df)
plt.title('Rating Distribution by Price Range')
plt.show()
```

#### Category Performance

```python
category_ratings = df.groupby('category')['rating'].mean().sort_values(ascending=False).head(15)
plt.figure(figsize=(7, 5))
sns.barplot(y=category_ratings.index, x=category_ratings.values, hue=category_ratings.index, palette='coolwarm', legend=False)
plt.xlabel("Average Rating")
plt.ylabel("Category")
plt.title("Category Performance by Average Rating")
plt.show()
```

#### Most Popular Product by Review Count


```python
popular_products = df.groupby('product_name')['rating_count'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
sns.barplot(y=popular_products.index, x=popular_products.values, hue=popular_products.index, palette='magma', legend=False)
plt.xlabel("Number of Reviews")
plt.ylabel("Product Name")
plt.title("Top 10 Most Reviewed Products")
plt.show()
```
